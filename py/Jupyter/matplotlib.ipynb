```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 2, 6, 4, 5])
y1 = np.array([2, 4, 6, 7, 19])

# con questa notazione possiamo separare i due oggetti che constituiscono il grafico
# fig è l'intero grafico, ax (sta per axes) contiene gli asxis
fig, ax = plt.subplots()

# possiamo assegnare dei label
ax.plot(x, y, label="prima func") # assegnamo ad ax (axes) i due singoi assi x, y
ax.plot(x, y1, label="seconda func") # possiamo disegnare più funzioni contemporaneamente

# bisogna chiamare la legenda
ax.legend()

# possiamo impostare una griglia
ax.grid()

# possiamo dare dei nomi agli assi (volendo a tutti i valori presenti sui due assi)
ax.set_xlabel("asse x")
ax.set_ylabel("asse y")

# possiamo impostare un titolo
ax.set_title("Il mio primo grafico") # come i bambini, i primi passi

plt.show()

```

```python
# possiamo anche parametrizzare i .subplots()
fig2, (ax1, ax2) = plt.subplots(2, 1)
ax1.plot(x, y)
ax2.plot(x, y1)

# possiamo inserire dei valori min, max per x, y (in questo caso solo per y)
ax1.set_ylim(0, 20)
ax2.set_ylim(0, 20)

plt.show()
```

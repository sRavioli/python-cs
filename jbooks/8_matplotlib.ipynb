Importiamo `matplotlib` e `numpy`

```python
import matplotlib.pyplot as plt
import numpy as np
```

Generiamo un range di numeri semi-casuale con un seed

```python
rng = np.random.default_rng(42)
```

Usiamo questo codice per generare un insieme di coordinate $x$ e $y$ che saranno
uniti a schermo mediante una spezzata:

```python
def plot_xy():
    x = np.arange(1, 6)
    y = rng.integers(low=0, high=10, size=5)
    
    # `figure` rappresenta l'intero grafico, `axes` contiene i singoli `axis`
    figure, axes = plt.subplots()
    axes.plot(x, y)

    return figure, axes

figure, axes = plot_xy()
```

Possiamo dare un nome agli assi e alla figura, impostando una griglia di riferimento.

```python
figure, axes = plot_xy()

# imposta un titolo
axes.set_title("Primo plot")

# imposta dei nomi per gli assi
axes.set_xlabel("asse x")
axes.set_ylabel("asse y")

# rende visibile la griglia
axes.grid(True)

plt.show()
```

È possibile impostare la griglia di riferimento su un unico asse specificando il
parametro `axis`. Sottolineamo come valorizzandolo a `"x"` sarà generata soltanto
la griglia per i valori di $x$ (ottenendo un reticolato verticale e parallelo 
all'asse delle ordinate), mentre viceversa con la valorizzazione a `"y"` sarà
generata la griglia per i valori di $y$ (ottenendo un reticolato orizzontale e
parallelo all'asse delle ascisse).

```python
figure, axes = plot_xy()

# rende visibile la griglia solo sull'asse x
axes.grid(axis="x")

plt.show()
```

**Esempio 1:** plot di più funzioni

Proviamo a tracciare due funzioni, una retta e un seno.

```python
# imposta dei valori per l'asse x e i due assi y
x = np.arange(0.0, 10.0, 0.01)
y1 = 2*x + 1
y2 = np.sin(x)

# inizializza
figure, axes = plt.subplots()

# traccia i grafici e imposta nomi degli stessi (per legenda)
axes.plot(x, y1, label="retta")
axes.plot(x, y2, label="funzione seno")

# imposta labels
axes.set_title("Due funzioni matematiche")
axes.set_xlabel("asse x")
axes.set_ylabel("asse y")

# mostra legenda e griglia
axes.legend()
axes.grid()   # True può essere omesso

plt.show()
```

**Esempio 2:** `plt.subplots()`

Proviamo ad organizzare le funzioni che abbiamo tracciato in precedenza in due
diversi `subplots()`.

```python
# la funzione `subplots()` accetta dei parametri, ovvero la disposizione dei
# subplots. in questo caso 2 righe, 1 colonna
figure, (axes1, axes2) = plt.subplots(2, 1)

# è possibile disegnare 4 grafici (o più)
# figure, (axis1, axis2) = plt.subplots(2, 2)

# imposta titolo
figure.suptitle("Due subplot di due funzioni matematiche")

# primo grafico
axes1.plot(x, y1, label="retta")
axes1.set_ylabel("asse y")
axes1.legend()
axes1.grid(True)

# secondo grafico
axes2.plot(x, y2, label="funzione seno")
axes2.set_xlabel("asse x")
axes2.set_ylabel("asse y")
axes2.legend()
axes2.grid(True)

# mostra la figura
plt.show()
```

**Esempio 3:** istogrammi

Proviamo a visualizzare a schermo un istogramma.

```python
x = rng.integers(low=0, high=100, size=1000)

figure, axes = plt.subplots()

axes.hist(x, edgecolor='black', linewidth=1.2, density=True)

axes.set_xlabel('bin')
axes.set_ylabel('conteggio dei singoli elementi')
axes.set_title('esempio di istogramma')

# salva la figura
# plt.savefig('hist.png')

plt.show()
```
